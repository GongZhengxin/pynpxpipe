```python
import spikeinterface.core as sc
import spikeinterface.postprocessing as sp

# 假设你已经有了原始的 recording 对象和同步后的 sorting_synced 对象
# recording, sorting_synced

# 1. 创建一个 WaveformExtractor 对象，这是进行后处理的中心
# 它会提取每个 spike 周围的波形片段并存储
we = sc.extract_waveforms(recording, sorting_ks4, folder=r'F:\ProcessPipeline\testdata\wordfob\kilosort_def_5block_97\sorter_output\waveforms_cache',
                           ms_before=3, ms_after=4, overwrite=True )

# 2. 现在，你可以轻松地获取 Matlab 循环中聚合的所有信息
for unit_id in we.unit_ids:
    print(f"\n--- Processing Unit {unit_id} ---")

    # 提取平均波形 (get_template)
    waveform = we.get_template(unit_id=unit_id)
    print(f"Waveform shape: {waveform.shape}")

    # 提取发放时间 (get_unit_spike_train - 已在上面演示)
    spike_train = sorting_ks4.get_unit_spike_train(unit_id=unit_id, time_unit='ms')
    print(f"Number of spikes: {len(spike_train)}")

    # 计算平均位置 (compute_unit_locations)
    # 这个函数会完成 Matlab 代码中的筛选和平均操作
    unit_locations = sp.compute_unit_locations(we)
    unit_pos = unit_locations[we.unit_ids.tolist().index(unit_id)]
    print(f"Unit position: {unit_pos}")

    # 提取振幅 (compute_spike_amplitudes)
    # 这个函数会计算每个 spike 的振幅
    amplitudes = sp.compute_spike_amplitudes(we)
    unit_amps = amplitudes[0][we.sorting.get_unit_spike_train(unit_id, segment_index=0)]
    # 注意: 上面的索引方式有点复杂，更简单的方式是直接获取
    # unit_amps = we.get_amplitudes(unit_id) # 需要先计算并保存
    print(f"Number of amplitudes: {len(unit_amps)}")
```

```python
from neuroconv.datainterfaces import SpikeGLXRecordingInterface
from neuroconv.utils.dict import load_dict_from_file, dict_deep_update
from pathlib import Path
from dateutil import tz
import spikeinterface.exporters as sie 
import spikeinterface.full as si
import spikeinterface.extractors as se
from pynwb import NWBHDF5IO

data_path = Path(fr'F:\ProcessPipeline\testdata\wordfob\NPX_MD241029_exp_g0\NPX_MD241029_exp_g0_imec0')
file_path = 'NPX_MD241029_exp_g0_t0.imec0.ap.bin'
interface = SpikeGLXRecordingInterface(file_path=str(data_path/file_path))

metadata = interface.get_metadata()
# assign local timezone
metadata["NWBFile"]["session_start_time"] = metadata["NWBFile"]["session_start_time"].replace(tzinfo=tz.tzlocal())

# session template from yaml
metadata_path = fr"F:\ProcessPipeline\pyneuralpipe\config\nwb_template.yaml"
metadata_from_yaml = load_dict_from_file(file_path=metadata_path)
# subject template from yaml
subject_path = fr"F:\ProcessPipeline\pyneuralpipe\config\FaLaDi.yaml"
subject_from_yaml = load_dict_from_file(file_path=subject_path)

# update metadata
metadata = dict_deep_update(metadata_from_yaml, metadata)
metadata = dict_deep_update(subject_from_yaml, metadata)

metadata["Ecephys"]["ElectrodeGroup"][0]["location"] = "MLO"
nwbfile = interface.create_nwbfile(metadata=metadata)
import os
ks_dir = fr"F:\ProcessPipeline\testdata\wordfob\kilosort_def_5block_97\sorter_output"

# 1) 读取 Kilosort4 排序结果
# 大多数版本可用：自动检测KS版本
sorting = se.read_kilosort(ks_dir)
recording = se.read_spikeglx(data_path, stream_name='imec0.ap')
# 2) 可选：先计算波形（推荐）
# 这样导出时会把均值/标准差波形一并写入 Units
waveforms_dir = os.path.join(ks_dir, "waveforms_cache")
analyzer = si.create_sorting_analyzer(
    sorting=sorting,
    recording=recording,
    folder=waveforms_dir,
    format="binary_folder",
    sparse=True,
    overwrite=True  # 或 load_if_exists=True
)

# 计算波形（可选其他扩展）
analyzer.compute(["random_spikes", "waveforms"], 
                random_spikes={"max_spikes_per_unit": 300},
                waveforms={"ms_before": 1.0, 
                          "ms_after": 2.0,
                          "return_scaled": True,
                          "n_jobs": 8})

# 修正：使用正确的导入
sie.export_to_nwb(
    sorting=sorting,
    recording=None,  # 建议提供以确保电极表正确
    sorting_analyzer=analyzer,  # 使用 sorting_analyzer 而非 waveforms
    nwbfile=nwbfile
)

from neuroconv.tools import configure_and_write_nwbfile
# final save

nwb_output_path = fr"F:\ProcessPipeline\nwbfiles\Test_20250924.nwb"
configure_and_write_nwbfile(
    nwbfile, nwbfile_path=nwb_output_path, backend="hdf5"
)
```
